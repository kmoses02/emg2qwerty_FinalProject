**Plots**

```python
import glob
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from omegaconf import OmegaConf

# Find latest run directory inside outputs-GRU
run_dirs = glob.glob("/content/drive/MyDrive/C247A/C247A-Final_Project/outputs-GRU/*/*")
assert run_dirs, "No run directories found in outputs-GRU."
latest_run = max(run_dirs, key=os.path.getmtime)
print("Using run directory:", latest_run)

# Find metrics.csv
metrics_files = glob.glob(os.path.join(latest_run, "**", "metrics.csv"), recursive=True)
assert metrics_files, "No metrics.csv found in latest run."
metrics_file = metrics_files[0]
print("Using metrics file:", metrics_file)

# Find saved Hydra config
config_files = glob.glob(os.path.join(latest_run, "**", "base.yaml"), recursive=True) + \
               glob.glob(os.path.join(latest_run, "**", "*.yaml"), recursive=True)

cfg = None
for f in config_files:
    try:
        maybe_cfg = OmegaConf.load(f)
        if "model" in maybe_cfg:
            cfg = maybe_cfg
            print("Using config file:", f)
            break
    except Exception:
        pass

assert cfg is not None, "Could not find a readable Hydra config with model settings."

# Read metrics
df = pd.read_csv(metrics_file)

# Extract CER rows
cer_df = df[["epoch", "val/CER"]].dropna()

# Pull hyperparameters from config
model_cfg = cfg.model
trainer_cfg = cfg.trainer

model_name = model_cfg.get("_target_", "unknown").split(".")[-1]
hidden_size = model_cfg.get("hidden_size", "n/a")
num_layers = model_cfg.get("num_layers", "n/a")
dropout = model_cfg.get("dropout", "n/a")
bidirectional = model_cfg.get("bidirectional", "n/a")
in_features = model_cfg.get("in_features", "n/a")
mlp_features = model_cfg.get("mlp_features", "n/a")

batch_size = cfg.get("batch_size", "n/a")
lr = cfg.optimizer.get("lr", "n/a") if "optimizer" in cfg else "n/a"
max_epochs = trainer_cfg.get("max_epochs", "n/a")

# Build text box
hyper_text = (
    f"Model: {model_name}\n"
    f"in_features: {in_features}\n"
    f"mlp_features: {mlp_features}\n"
    f"hidden_size: {hidden_size}\n"
    f"num_layers: {num_layers}\n"
    f"dropout: {dropout}\n"
    f"bidirectional: {bidirectional}\n"
    f"batch_size: {batch_size}\n"
    f"lr: {lr}\n"
    f"max_epochs: {max_epochs}\n"
    f"run: {Path(latest_run).name}"
)

# Plot CER
plt.figure(figsize=(10, 6))
plt.plot(cer_df["epoch"], cer_df["val/CER"], marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation CER")
plt.title(f"Validation CER vs Epoch ({model_name})")
plt.grid(True)

# Add hyperparameter box
plt.gcf().text(
    0.72, 0.5, hyper_text,
    fontsize=10,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="gray")
)

plt.tight_layout(rect=[0, 0, 0.72, 1])
plt.show()

print("Best val/CER:", cer_df["val/CER"].min())
```

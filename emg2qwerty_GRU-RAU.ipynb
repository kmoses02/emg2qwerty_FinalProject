Define run metadata

```python
from datetime import datetime

# Hyperparameters for this run
HIDDEN_SIZE = 512                 # Try 128 / 256 / 512 / 768
NUM_LAYERS = 2                    # Try 1 / 2 / 3
DROPOUT = 0.1                     # Try 0.1 / 0.2 / 0.3
BIDIRECTIONAL = True
LR = 0.001                        # Try 0.0003 / 0.0005 / 0.001 / 0.003
BATCH_SIZE = 32                   # Try 16 / 32 / 64
MAX_EPOCHS = 50                    # Try 50  / 100 /

# Build short tags
drop_tag = str(DROPOUT).replace(".", "")
bi_tag = "bi" if BIDIRECTIONAL else "uni"
lr_tag = f"{LR:.0e}".replace("-", "m")   # 0.001 -> 1e-03 -> 1em03

# Automatic run name
# RUN_NAME = f"gru_h{HIDDEN_SIZE}_l{NUM_LAYERS}_d{drop_tag}_{bi_tag}_lr{lr_tag}_b{BATCH_SIZE}_e{MAX_EPOCHS}"
RUN_NAME = f"gru_noaug_h{HIDDEN_SIZE}_l{NUM_LAYERS}_d{drop_tag}_{bi_tag}_lr{lr_tag}_b{BATCH_SIZE}_e{MAX_EPOCHS}"

timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
RUN_DIR = f"/content/drive/MyDrive/C247A/C247A-Final_Project/outputs-GRU-RAU/{timestamp}_{RUN_NAME}"

print("RUN_NAME:", RUN_NAME)
print("RUN_DIR :", RUN_DIR)
```

Load metrics and config

```python
import glob
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from omegaconf import OmegaConf

run_path = Path(RUN_DIR)
assert run_path.exists(), f"Run directory not found: {RUN_DIR}"

metrics_files = glob.glob(str(run_path / "csv_logs" / "**" / "metrics.csv"), recursive=True)
assert metrics_files, f"No metrics.csv found under {run_path / 'csv_logs'}"

metrics_file = metrics_files[0]

config_files = glob.glob(str(run_path / "**" / "*.yaml"), recursive=True)
cfg = None
cfg_path = None
for f in config_files:
    try:
        maybe_cfg = OmegaConf.load(f)
        if "model" in maybe_cfg:
            cfg = maybe_cfg
            cfg_path = f
            break
    except Exception:
        pass

assert cfg is not None, "Could not find config with model settings."

df = pd.read_csv(metrics_file)

print("Using metrics file:", metrics_file)
print("Using config file:", cfg_path)
print("Columns:", df.columns.tolist())
```

Save summary JSON

```python
model_cfg = cfg.model
trainer_cfg = cfg.trainer

summary = {
    "run_name": RUN_NAME,
    "run_dir": RUN_DIR,
    "model_target": model_cfg.get("_target_", "unknown"),
    "in_features": model_cfg.get("in_features", "n/a"),
    "mlp_features": list(model_cfg.get("mlp_features", [])),
    "hidden_size": model_cfg.get("hidden_size", "n/a"),
    "num_layers": model_cfg.get("num_layers", "n/a"),
    "dropout": model_cfg.get("dropout", "n/a"),
    "bidirectional": model_cfg.get("bidirectional", "n/a"),
    "batch_size": cfg.get("batch_size", "n/a"),
    "lr": cfg.optimizer.get("lr", "n/a") if "optimizer" in cfg else "n/a",
    "max_epochs": trainer_cfg.get("max_epochs", "n/a"),
}

if "val/CER" in df.columns:
    vals = df["val/CER"].dropna()
    if len(vals) > 0:
        summary["best_val_CER"] = float(vals.min())

if "test/CER" in df.columns:
    vals = df["test/CER"].dropna()
    if len(vals) > 0:
        summary["final_test_CER"] = float(vals.iloc[-1])

with open(run_path / "summary.json", "w") as f:
    json.dump(summary, f, indent=2)

print("Saved:", run_path / "summary.json")
print(summary)
```

**CER plot**

```python
cer_df = df[["epoch", "val/CER"]].dropna()

hyper_text = (
    f"Model: {summary['model_target'].split('.')[-1]}\n"
    f"in_features: {summary['in_features']}\n"
    f"mlp_features: {summary['mlp_features']}\n"
    f"hidden_size: {summary['hidden_size']}\n"
    f"num_layers: {summary['num_layers']}\n"
    f"dropout: {summary['dropout']}\n"
    f"bidirectional: {summary['bidirectional']}\n"
    f"batch_size: {summary['batch_size']}\n"
    f"lr: {summary['lr']}\n"
    f"max_epochs: {summary['max_epochs']}\n"
    f"run: {run_path.name}"
)

plt.figure(figsize=(10, 6))
plt.plot(cer_df["epoch"], cer_df["val/CER"], marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation CER")
plt.title("Validation CER vs Epoch")
plt.grid(True)

plt.gcf().text(
    0.72, 0.5, hyper_text,
    fontsize=10,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="gray")
)

plt.tight_layout(rect=[0, 0, 0.72, 1])

cer_plot_path = run_path / "val_cer_plot.png"
plt.savefig(cer_plot_path, dpi=200, bbox_inches="tight")
plt.show()

print("Saved plot to:", cer_plot_path)
print("Best val/CER:", summary.get("best_val_CER", "n/a"))
print("Final test/CER:", summary.get("final_test_CER", "n/a"))
```
